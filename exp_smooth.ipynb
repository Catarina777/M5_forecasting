```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Reading in the data
data = pd.read_csv('df_ca_new.csv')
```

```python
data.drop('Unnamed: 0', axis=1, inplace=True)
data
```

```python
# Checking for NaN values in each column
nan_columns = data.columns[data.isna().any()].tolist()

print("Columns with NaN values:", nan_columns)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.holtwinters import SimpleExpSmoothing
from sklearn.metrics import mean_squared_error

# Prepare data with 'sold' column
predict = data[['sold']]

# Train-test split
train, test = predict.iloc[:-28], predict.iloc[-28:]

# Fit Simple Exponential Smoothing model
model = SimpleExpSmoothing(train)
ses_model = model.fit()

# Forecast the next 28 days
forecast = ses_model.forecast(steps=28)

# Ensure no NaN values in forecast
if np.isnan(forecast).any():
    print("Warning: NaN values found in forecast. Filling with zeros.")
    forecast = forecast.fillna(0)

# Calculate RMSE
rmse = np.sqrt(mean_squared_error(test, forecast))
print("RMSE:", rmse)

# Generate date range for forecast period
forecast_dates = pd.date_range(start=test.index[0], periods=28, freq='D')

# Slice actual sales data to match the length of the forecast
actual_sales = test['sold'].iloc[:len(forecast)]

# Visualize results
plt.plot(forecast_dates, actual_sales, label='Actual')
plt.plot(forecast_dates, forecast, label='Forecast')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.title('Actual vs Forecasted Sales')
plt.legend()
plt.show()
```
